```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census

# Census API Key
from census_key import census_key
c = Census(census_key, year=2012)
```

```python
census_data = c.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
sandy_census_pd = pd.DataFrame(census_data)

sandy_census_pd = sandy_census_pd.rename(columns={"B25035_001E": "Median House Consturction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
sandy_census_pd.head()
```

```python
ac_df = sandy_census_pd.loc[sandy_census_pd["Name"] == "Atlantic County, New Jersey"]

ac_df
```
